```python
import regex

with open('day1/input1.txt') as f:
    lines = f.readlines()
```

```python
# part 1
total = 0
for el in lines:
    el1 = regex.search("\d", el)[0]
    el2 = regex.search("(\d)(\D*$)", el)[1]
    total += int(el1+el2)
total
```

```python
# part 2
conv_dict = {'one':1, 
             'two':2, 
             'three':3, 
             'four':4,
             'five':5, 
             'six':6, 
             'seven':7,
             'eight':8,
             'nine':9, 
            'eno':1, 
             'owt':2, 
             'eerht':3, 
             'ruof':4,
             'evif':5, 
             'xis':6, 
             'neves':7,
             'thgie':8,
             'enin':9}


total = 0
for el in lines:
    el1 = regex.search("\d|one|two|three|four|five|six|seven|eight|nine", el)[0]
    if el1 in conv_dict.keys(): 
        el1 = conv_dict[el1]
    el2 = regex.search("\d|eno|owt|eerht|ruof|evif|xis|neves|thgie|enin", el[::-1])[0]
    if el2 in conv_dict.keys(): 
        el2 = conv_dict[el2]
    total += int(str(el1)+str(el2))
total
```
